# Julia 基礎圖表繪製: Plots 套件

## 作業 023：繪製 3D 函數

練習繪製 3D 函數：$z = x^2 - y^2$，$x$ 範圍於 $-5 \sim 5$，$y$ 範圍於 $-5 \sim 5$ 之間。（請完成以下程式碼）

```julia
using Plots
```

```julia
n = 3000
x = range(-5., stop=5., length=n)
y = range(-5., stop=5., length=n)
```

```julia
z = f(x,y) = (x ^ 2 - y ^ 2) 
```

```julia
p = plot(x,y,z,st=:surface)
```
